```python
## This piece of code reads a file in CSV format, converts to gml format
# Reference: https://gist.github.com/dsignr/c3f7a67fcfb1fb93698a507f4cce8eef#file-dataframe-to-gml-ipynb
# This code is overall generic with capability to handle the CSV dataset in scope
import sys, time
import pandas as pd
import datetime as dt
from IPython.display import display



FILE_NAME = "D:\\IU\\1_Network_Science\\Project\\CSV_GML\\2016_aggregated_v5.gml"


df = pd.read_csv('D:\\IU\\1_Network_Science\\Project\\CSV_GML\\2016_aggregated_v3.csv', sep=',')
cols = ['ORIGIN_AIRPORT_ID','DEST_AIRPORT_ID','TOTAL_PASSENGER']
df[cols] = df[cols].astype(int)

print(df.head(5))

#Utility functions
def progress(v):
    v = str(v)
    sys.stdout.flush()
    sys.stdout.write('\r')
    sys.stdout.flush()
    sys.stdout.write(v)
    
```

```python
f = open(FILE_NAME, "w")
#helpers
s = " "
ss = s+s
sss = s+s+s
ssss = s+s+s+s
nl = "\n"

#loop helpers
added = []
ind = 0

#Root node
f.write("graph"+nl)
f.write("["+nl)

##Switch the lines below to create 
## Directed or Undirected graph
#Mark the network directed
#f.write( ss + "directed 0" + nl)
#Mark the network undirected
f.write( ss + "directed 1" + nl)

#Write an edge
def write_edge(r):
    f.write( ss + "edge" + nl)
    f.write( ss + "[" + nl)
    #f.write( ssss + "source" + s + '"' + str(r['ORIGIN_ID']) + '"' + nl)
    #f.write( ssss + "target" + s + '"' + str(r['DEST_ID']) + '"' + nl)
    #f.write( ssss + "value" + s + str(r['PASSENGERS']) + '"' + nl)
    f.write( ssss + "source" + s + str(r['ORIGIN_AIRPORT_ID']) + nl)
    f.write( ssss + "target" + s + str(r['DEST_AIRPORT_ID']) + nl)
    ####f.write( ssss + "value" + s + str(r['TOTAL_PASSENGER']) + nl)
    f.write( ss + "]"+ nl)
    
#Write a node
def write_node_ORIGIN(r):
    f.write( ss + "node" + nl)
    f.write( ss + "[" + nl)
    #f.write( ssss + "id" + s + '"' + str(r['ORIGIN_ID']) + '"' + nl)
    f.write( ssss + "id" + s + str(r['ORIGIN_AIRPORT_ID']) + nl)
    f.write( ssss + "label" + s + '"' + str(r['ORIGIN']) + '"' + nl)
    f.write( ss + "]"+ nl)
    
def write_node_DEST(r):
    f.write( ss + "node" + nl)
    f.write( ss + "[" + nl)
    #f.write( ssss + "id" + s + '"' + str(r['ORIGIN_ID']) + '"' + nl)
    f.write( ssss + "id" + s + str(r['DEST_AIRPORT_ID']) + nl)
    f.write( ssss + "label" + s + '"' + str(r['DEST']) + '"' + nl)
    f.write( ss + "]"+ nl)
    
    
#Generate nodes
for i, r in df.iterrows():
    #increment, as index not reliable
    ind += 1
    #Check for duplicates
    if (r['ORIGIN_AIRPORT_ID'] not in added):
        #Add to list
        added.append(r['ORIGIN_AIRPORT_ID'])
        write_node_ORIGIN(r)
        
    if (r['DEST_AIRPORT_ID'] not in added):
        #Add to list
        added.append(r['DEST_AIRPORT_ID'])
        write_node_DEST(r) 
            
    #print the progress    
    progress(ind)

print(nl+"Printing nodes over")

#flush index
ind = 0    
#Generate edges    
for i, r in df.iterrows():
    #increment, as index not reliable
    ind += 1
    #if(r['ORIGIN_ID'] < r['DEST_ID']):
     #   write_edge(r)
    write_edge(r)
    #print the progress            
    progress(ind)
    
print(nl+"Printing nodes and edges over")

#closing node
f.write("]"+nl)
f.close()
```
